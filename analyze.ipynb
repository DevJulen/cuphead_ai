# Analyzing the results

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('results.csv', skipinitialspace=True)
df
```

- Box loss through the epochs on the training and validation sets

```python
plt.plot(df['epoch'], df['train/box_loss'])
plt.plot(df['epoch'], df['val/box_loss'])
plt.legend(["Train", "Val"])
plt.xlabel("Epochs")
plt.ylabel("Box loss")
plt.show()
```

- Object loss through the epochs on the training and validation sets

```python
plt.plot(df['epoch'], df['train/obj_loss'])
plt.plot(df['epoch'], df['val/obj_loss'])
plt.legend(["Train", "Val"])
plt.xlabel("Epochs")
plt.ylabel("Object loss")
plt.show()
```

- Class loss through the epochs on the training and validation sets

```python
plt.plot(df['epoch'], df['train/cls_loss'])
plt.plot(df['epoch'], df['val/cls_loss'])
plt.legend(["Train", "Val"])
plt.xlabel("Epochs")
plt.ylabel("Class loss")
plt.show()
```

- mAP with a 0.5 IOU threshold through the epochs on the validation set

```python
plt.plot(df['epoch'], df['metrics/mAP_0.5'])
plt.xlabel("Epochs")
plt.ylabel("mAP_0.5")
plt.show()
```

- mAP with an IOU threshold betwen 0.5:0.95 through the epochs on the validation set

```python
plt.plot(df['epoch'], df['metrics/mAP_0.5:0.95'])
plt.xlabel("Epochs")
plt.ylabel("mAP_0.5:0.95")
plt.show()
```

- Precission through the epochs on the validation set

```python
plt.plot(df['epoch'], df['metrics/precision'])
plt.xlabel("Epochs")
plt.ylabel("Precission")
plt.show()
```

- Precission through the epochs on the validation set

```python
plt.plot(df['epoch'], df['metrics/recall'])
plt.xlabel("Epochs")
plt.ylabel("Recall")
plt.show()
```
